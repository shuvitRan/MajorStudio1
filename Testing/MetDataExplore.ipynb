```python
def FindPainting(findplace):
    mydata= data[((data["Artist Nationality"].str.contains(findplace)==True)|
                 (data["Artist Display Name"].str.contains(findplace)==True)|
                 (data["Artist Display Bio"].str.contains(findplace)==True)|
                 (data.Culture.str.contains(findplace)==True)|
                 (data.Country.str.contains(findplace)==True))
                &
                ((data.Classification.str.contains('Painting|Codices', case = False)==True)|
#                 (data.Classification.str.contains('Drawing', case = False)==True)|
                (data["Artist Prefix"].str.contains('Drawing|Painting', case = False)==True))]
    mydata["Tags"] = mydata["Tags"].str.split("|")
#     mydata["Tags"].str.split("|")
#     mydata.apply(mydata["Tags"].str.split("|"))
    return mydata
```

```python
print('USA '+ str(len(FindPainting("United States|American"))))
print('German '+ str(len(FindPainting("German|Germany"))))
print('Japan '+ str(len(FindPainting("Japanese|Japan"))))
print('UK '+ str(len(FindPainting("British|England|United Kindom"))))
print('French '+ str(len(FindPainting("French|France"))))
print('China '+ str(len(FindPainting("China|Chinese"))))
print('Iran '+ str(len(FindPainting("Iran|Iranian"))))
print('Italy '+ str(len(FindPainting("Italy|Italian"))))
print('India '+ str(len(FindPainting("India|Indian"))))
```
